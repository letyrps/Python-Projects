Ánalise de tipos de erros:

Isolando valores não nulos:

```python
df_mambu_bool = df['resposta_mambu'].notna()
```

```python
df_mambu = df[df_mambu_bool]
```

```python
df_mambu
```

Separando informações da coluna resposta mambu e criando coluna de tipo de erro:

```python
error_reason_list = []
for value in df_mambu['resposta_mambu']:
    error_reason_all = value[value.find('errorReason='):]
    error_reason_all_list = error_reason_all.split(',')
    error_reason = error_reason_all_list[0]
    error_reason_list.append(error_reason)

error_reason_list
```

```python
df_mambu.insert(21, 'Error_Reason',error_reason_list)
```

Nova df:

```python
df_mambu
```

```python
df_mambu['Error_Reason'].value_counts()
```
